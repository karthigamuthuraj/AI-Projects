```python
import pandas as pd
df = pd.read_csv('whatsapp-chat-data.txt',header=None,on_bad_lines='skip',encoding='utf8')
```

```python
df
```

```python

# Concatenate values from columns 1, 2, and 3
df['Concatenated'] = df[[1, 2, 3]].apply(lambda row: ''.join(row.values.astype(str)), axis=1)
# Drop columns 1, 2, and 3
df = df.drop(0)
df = df.drop(columns=[1, 2, 3])

df.head(5)
```

```python
df.columns = ["Date","Chat"]
```

```python
df.head(10)
```

```python
Message = df["Chat"].str.split('-',n=1,expand=True)
Message1 = Message[1].str.split(':',n=1,expand=True)
df["Time"] = Message[0]
df["Name"] = Message1[0]
df["Chat"] = Message1[1]
df = df[["Date","Time","Name","Chat"]]
```

```python

df.info()
```

```python
import nltk

nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
dataset = df 
```

```python
dataset.dropna(inplace=True)
```

```python
dataset
```

```python
sid = SentimentIntensityAnalyzer()

senti = sid.polarity_scores(dataset["Chat"][114])
senti
```

```python
# Apply sentiment analysis to each message
df[['Neg', 'Neu', 'Pos', 'Compound']] = df['Chat'].apply(lambda x: pd.Series(sid.polarity_scores(str(x))))

# Display the first few rows with sentiment scores
print(df.head())
```

```python
# Calculate sentiment scores for each message
sentiment_scores = df['Chat'].apply(lambda x: sid.polarity_scores(str(x)))
# Extract the compound score from the sentiment analysis
df['Compound'] = sentiment_scores.apply(lambda score_dict: score_dict['compound'])

# Classify messages as positive, negative, or neutral
df['Sentiment'] = df['Compound'].apply(lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))

# Reorder columns to include sentiment scores
df = df[["Date", "Time", "Name", "Chat","Neg", "Neu", "Pos",  "Compound", "Sentiment"]]

# Save the DataFrame to a CSV file
df.to_csv('whatsapp_chat_sentiment_analysis.csv', index=False)
```
